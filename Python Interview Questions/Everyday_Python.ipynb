### problem 1 :

The Challenge:
Write a Python function that takes a string as input and returns a dictionary. The dictionary keys should be the characters in the string, and the values should be the number of times each character appears.

Example:

Input: "banana"

Output: {'b': 1, 'a': 3, 'n': 2}

```python
li = "banana"
dic = dict.fromkeys(li,0)
# print(dic.values())

for char in li:
    if char in dic:
        dic[char] += 1
print(dic)
        
```

### Problem- 2
The "FizzBuzz" List Filter
The Challenge:
Write a Python program that takes a list of numbers from 1 to 20. Create a new list based on these rules:

If the number is divisible by 3, replace it with the string "Fizz".

If the number is divisible by 5, replace it with the string "Buzz".

If it is divisible by both 3 and 5, replace it with "FizzBuzz".

Otherwise, just keep the number.

Example:

Input: [1, 2, 3, 4, 5, 15]

Output: [1, 2, "Fizz", 4, "Buzz", "FizzBuzz"]

```python
p = [1,2,3,10,15]
for n,i in enumerate(p):
    if i%3==0 and i%5==0:
        p[n] = "FizzBuzz"
    elif i%3 ==0: 
        p[n] = "Fizz"
    elif  i%5 ==0:
        p[n] = "Buzz"     
print(p)   
```

The List "Unique-ifier"
Let's stick with the challenge I mentioned before, as it introduces a very powerful Python concept.

The Challenge:
You have a list with many duplicate items. You need to remove all duplicates so that every item appears only once.

Example:

Input: ["apple", "banana", "apple", "orange", "banana", "apple"]

Output: ["apple", "banana", "orange"] (The order doesn't strictly matter for this challenge).

```python
Input = ["apple", "banana", "apple", "orange", "banana", "apple"]
p = []
for i in Input:
    if i not in p:
        p.append(i)
print(p)

######## Way 2####

set(Input)
        
```

```python
Input = "programming is fun"
vovel = "aeiou"
dic= dict.fromkeys(vovel,0)

for char in Input:
    lw = char.lower()
    if lw in vovel:
        dic[lw] +=1 
print(dic)
    
```

Check Palindrom for number

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if str(x) == str(x)[::-1]:
            return True
        else:
            return False
        
```

```python
def fibonacci(n):
    fib_sequence = [0, 1]  # Start with the base cases: Fib(0) = 0, Fib(1) = 1

    # Generate Fibonacci sequence up to the nth term
    for i in range(2, n + 1):
        next_fib = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_fib)

    return fib_sequence[:n]

# Example usage:
n = 10  # Number of Fibonacci numbers to generate
fib_sequence = fibonacci(n)
print(f"Fibonacci sequence up to {n} terms:", fib_sequence)



# FIbonacchi

def fib_seq(n):
  fib_ = [1,2]

  for i in range(1,n+1):
    next_fib = fib_[-1] + fib_[-2]
    fib_.append(next_fib)
  return fib_

fib_seq(10)
```

```python
# armstrong number
# ex. = 153 = (1*1*1)+(5*5*5)+(3*3*3)
#           = 153
# ex. = 52 = (5*5) + (2*2)
#          = 52

# prompt: write program to check number is armstrong number

# Python program to check if the number is Armstrong number or not

# Function to check if the number is Armstrong number or not
def isArmstrong(num):
    # convert the number to string
    str_num = str(num)

    # find the length of the number
    length = len(str_num)

    # calculate the sum of the power of each digit
    sum = 0
    for i in range(0, length):
        sum += int(str_num[i]) ** length

    # check if the sum is equal to the number
    if sum == num:
        return True
    else:
        return False

# take input from the user
num = int(input("Enter a number: "))

# check if the number is Armstrong number or not
if isArmstrong(num):
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

```python
# Find HCF(Highest common factor) and GCD (Gretest Common Divisor)

def findHCF(x,y):


  if x > y :
    smaller = y
  else:
    smaller = x

    for i in range(1,smaller+1):

      if((x%i ==0) and (y%i==0)):
        hcf = i
    return hcf

findHCF(12,24)
```

```python

# Write a Python function that:

# Calculates the total revenue for each customer.
# Returns a DataFrame with Customer_ID and Total_Revenue columns sorted in descending order of Total_Revenue.


import pandas as pd

# Sample data
data = {
    "Order_ID": [1, 2],
    "Customer_ID": [101, 102],
    "Order_Date": ["2024-11-01", "2024-11-02"],
    "Product": ["A", "B"],
    "Quantity": [3, 2],
    "Price": [50, 100]
}

# Create DataFrame
orders = pd.DataFrame(data)

# Function to calculate total revenue for each customer
def calculate_total_revenue(df):
    # Add a Total column for each order
    df['Total'] = df['Quantity'] * df['Price']
    # Group by Customer_ID and calculate Total Revenue
    total_revenue = df.groupby('Customer_ID')['Total'].sum().reset_index()
    # Rename column
    total_revenue.columns = ['Customer_ID', 'Total_Revenue']
    # Sort by Total_Revenue in descending order
    total_revenue = total_revenue.sort_values(by='Total_Revenue', ascending=False)
    return total_revenue

# Call the function
result = calculate_total_revenue(orders)
print(result)
```

```python

# Calculates the average salary of employees in each department.
# Returns a DataFrame with Department and Average_Salary, sorted alphabetically by Department.

import pandas as pd

# Sample data
data = {
    "Employee_ID": [1, 2, 3],
    "Department": ["HR", "IT", "IT"],
    "Joining_Date": ["2022-01-15", "2023-03-01", "2022-07-10"],
    "Salary": [50000, 70000, 80000]
}

# Create DataFrame
employees = pd.DataFrame(data)

# Function to calculate average salary per department
def calculate_average_salary(df):
    avg_salary = df.groupby('Department')['Salary'].mean().reset_index()
    avg_salary.columns = ['Department', 'Average_Salary']
    avg_salary = avg_salary.sort_values(by='Department')
    return avg_salary

# Call the function
result = calculate_average_salary(employees)
print(result)
```

```python
# Filters transactions where the Amount is greater than 250.
# Adds a new column Discounted_Amount where every transaction gets a 10% discount applied.
# Returns the modified DataFrame.
data = {
    "Transaction_ID": [1, 2, 3],
    "Product": ["A", "B", "A"],
    "Customer_ID": [101, 102, 103],
    "Amount": [500, 200, 300],
    "Transaction_Date": ["2024-01-01", "2024-01-15", "2024-02-01"]
}

# Create DataFrame
sales = pd.DataFrame(data)

def filterdf():

  filter_df = sales[sales['Amount'] > 250]
  filter_df['Disc'] = filter_df['Amount'] * 0.1
  return filter_df

result = filterdf()
print(result)
```

```python
# Converts the Purchase_Date column to a datetime format.
# Creates a new column Purchase_Month with the month name (e.g., "January").
# Aggregates the total Amount spent per month and returns the result as a DataFrame.

# Sample data
data = {
    "Customer_ID": [101, 102, 103],
    "Purchase_Date": ["2023-12-15", "2024-01-10", "2024-02-05"],
    "Amount": [150, 200, 300]
}

# Create DataFrame
purchases = pd.DataFrame(data)


def pur():

  purchases['Purchase_Date']  = pd.to_datetime(purchases['Purchase_Date'])
  purchases['Purchase_month'] = purchases['Purchase_Date'].dt.month_name()
  total_amount = purchases.groupby('Purchase_month')['Amount'].sum().reset_index()
  return total_amount

result = pur()
print(result)
```

```python
# Leet code
# 1) Example 1:

# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)

        for i in range(n-1):
            for j in range(i+1,n):
                if nums[i] + nums[j] == target:
                    return[i,j]
        return []
```

```python
# Explain the difference between deepcopy and shallow copy.
# Shallow Copy: Copies the outer object, but inner objects are still references to the original.
# Deep Copy: Creates a complete independent copy of the entire object, including nested objects.

import copy

original = [[1, 2], [3, 4]]
shallow = copy.copy(original)
deep = copy.deepcopy(original)

shallow[0][0] = 99
print(original)  # Output: [[99, 2], [3, 4]] (shallow copy affects the original)

deep[0][0] = 77
print(original)  # Output: [[99, 2], [3, 4]] (deep copy does not affect the original)
```

```python
# How do you implement a generator to yield Fibonacci numbers?



def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print(list(fibonacci(5)))  # Output: [0, 1, 1, 2, 3]
```

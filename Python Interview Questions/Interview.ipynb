```python
# Find HCF : Highest common factor

def HCF(x,y):
    if x > y:
        smaller = x
    else:
        smaller = y
    for i in range(1,smaller+1):
        if ((x%i==0) and (y%i==0)):
            hcf = i
    return hcf

HCF(20,30)

def GCD(x:int , y: int) -> int:
    while y:
      x,y = y,x%y
    return x

GCD(20,30)
```

```python
def sanityCheck(data_type):
    def outer_function(func):
        def inner_function(*args):
            if type(args[0]) == data_type:
                func(*args)
            else:
                raise TypeError("wrong data type")
        return inner_function
    return outer_function


@sanityCheck(int)
def square(num):
    print(num**2)

square(2)
square("kalpesh")
```

```python
def fibb(n):
    a,b = 0,1
    for i in range(1,n+1):
        row = []
        
        for j in range(i):
            a,b = b,a+b
            row.append(a)
        print(" ".join(map(str,row)))
    
fibb(5)
```

```python
def find_prim(start,end):
    for num in range(start,end+1):
        if num >1:
            is_prime = True
            
            for i in range(2,int(num**0.5)+1):
                if num%i==0:
                    is_prime = False
                    break
            if is_prime:
                print(num)
```

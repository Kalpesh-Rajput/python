Constructure : __init__
whenever we crate/construct an object of a class, there is an inbuilt method (__init_) which is called known as constructor

```python
class student:
    def __init__(self,name,age,qf):
       self.name = name
       self.age = age
       self.qf = qf
    def study(self):
        print("padle " + self.name + " jake")
    def HW(self):
        print("jaldi se kar kutte teri age " + str(self.age)+" ho gai hai")
        
s1 = student("kalpesh",45,"msc")
s2 = student("Umesh",65,"bsc")
s1.study()
s2.HW()
```

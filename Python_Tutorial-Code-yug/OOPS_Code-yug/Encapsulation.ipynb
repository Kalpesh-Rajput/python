# Encpasulation

What is encapsulation in python ?
* Wraping the data and methods working on data together in a single unit(ex.class) is called as encapsulation

```python
# here we can see revenue is not private anyone can change the revenue to 0
class finance:
    def __init__(self):
        self.revenue = 50000
        self.sales = 104
f1 = finance()
print(f1.__dict__)
class hr:
    def __init__(self):
        self.num_of_emp = 32
        f1.revenue = 0
h1 = hr()
print(h1.__dict__)
print(f1.__dict__)
```

### Access Modifier in python :
* Generally we restrict data access outside the class in encapsulation
* encapsulation can achieved by declaring the data members and methods of a class as private
* three access specifier : 
* 1) Public : Accessible anywhere by using object reference  
* 2) Private: Accessible within class. accessible via methods only.  (usng = "__") to private
* 3) Protected : Accessible within class and it's subclasses (using = "_") to protected

```python
# here we can see revenue is not private anyone can change the revenue to 0
class finance:
    def __init__(self):
        self.__revenue = 50000
        self.sales = 104
f1 = finance()
print(f1.__dict__)
class hr:
    def __init__(self):
        self.num_of_emp = 32
        print(f1.__revenue)
h1 = hr()
print(h1.__dict__)
print(f1.__dict__)
```

```python
# Modification will not happen as we made it private
class finance:
    def __init__(self):
        self.__revenue = 50000
        self.sales = 104
f1 = finance()
print(f1.__dict__)
class hr:
    def __init__(self):
        self.num_of_emp = 32
        f1.__revenue = 15000 # Modification will not happen as we made it private
h1 = hr()
print(h1.__dict__)
print(f1.__dict__)
```

```python
# To access the private instance we use methods like getter to get the data 
class finance():
    def __init__(self):
        self.__revenue = 50000
        self.sales = 104
    def getrevenue(self):
        print(self.__revenue) # before modification
        self.__revenue = 5    # modification ahppens here
        print(self.__revenue) # after modification
        
f1  = finance()
print(f1.__dict__)  # still we can access revenue using dict but with different name format is "_classname__variablename" lets see can we access it using this name 
f1.getrevenue()
f1._finance__revenue # yes we can acces private variable using it so means in python no direct encapsulation is there we only use this to avoide changes directly
```

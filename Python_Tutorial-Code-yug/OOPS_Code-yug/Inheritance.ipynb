## Creating instance Variables :

```python
# Creating instance using constructer

class employee:
    def __init__(self,nm,m):  
        self.name = nm
        self.marks = m
        
e1 = employee("kalpesh",15)
e2 = employee("umesh",20)
print(e1.__dict__)
print(e2.__dict__)
```

```python
# Creating instance outside class
class student:
    def __init__(self,name,marks):
        self.name = name
        self.marks =marks
        
std1 = student("kalpesh"  , 95)
std2 = student("Umesh" , 85)
std2 = student("Rakesh" , 75)

# adding new instance
std1.age = 25
print(std1.__dict__)

# changing value of instance
std2.name = "rohan"
print(std2.__dict__)

#deleting instance
del std1.age

print(std1.__dict__)
```

```python
# Creating using instance method

class student:
    def __init__(self,name,marks):
        self.name = name
        self.marks =marks
    
    def display(self):
        print(self.name,self.marks)
        
    def change(self):
        self.name = input("enter name ")
        self.marks = int(input("enter age"))
        
std1 = student("kalpesh"  , 95)
std2 = student("Umesh" , 85)
std2 = student("Rakesh" , 75)

std1.display()

display(std1.__dict__)


std2.change()
std2.display()
```

# Class Methods : 
* method which work on class variable
* first argument is class reference
* made using decorator @classmethod

```python

class employee:
    company_name = "BCG"
    def __init__(self,nm,m):  
        self.name = nm
        self.marks = m
    @classmethod
    def get_company_name(cls):
        # print(f" company Name is :",cls.company_name)
        cls.company_name = "new game"
        print(cls.company_name)
        
employee.get_company_name()
    
```

# **Inheritance** :

Deriving a new class from an existing class so that new class inherits all members (attributes + methods) of existing class is called as inheritance

we have parent class and child class

class parent_class(object):
    
    attributes+ methods

class child_class(parent):

```python
class employee:
    bonus = 10000
    def display(self):
        print("employee class method")
        
class manager:
    bonus1 = 10000
    def show(self):
        print("manager class method")
        
e1 = employee()
m1 = manager()

e1.display()
m1.show()
# e1.show()
m1.bonus# got error cause we call manager method to employee class
```

```python
# Using Inheritance
class employee:
    bonus = 10000
    def display(self):
        print("employee class method")
        
class manager(employee):
    bonus1 = 10000
    def show(self):
        print("manager class method")
        
e1 = employee()
m1 = manager()

e1.display()
m1.show()
m1.bonus 
e1.bonus1 # will get error cause we can not access child class attrbtes in parent class
```

Super() function :-

* using super() function we can access parent clas properties
* this function return a temporary object which contain reference to parent class
* it makes inheritance more manageable and extensible

```python
# Now above code will run without error
class father:
    def __init__(self):
        print("father class constuctor")
        self.vehicle = "scooter"

class son(father):
    def __init__(self):
        super().__init__()
        print("son class constructor")
        self.vehicle1  = "R15"
        self.home = "5BHK"

s = son()
print(s.vehicle)
```

### Type of inheritance : Depending  on number of child and parent classes involved  
* Single  : One parent and One chind class (Object --> Parent --> Child)
* Multi-level : Parent class and clind class further inherited into new class forming multiple levels (Object --> parent --> Child --> Grand Child--> ...........)
* Hierachical  : One parent multiple chind ( Parent  --> child1 , child2 ,child3)
* multiple : Class derived from multiple base classes (Object --> Parent1 , Parent2 -- > Child)
* hybrid : It contain multiple type of inheritance (parent1,parent2 --> Child --> Child1 , Child2)
* Cyclic : same class inherite that same class means X class will inheritate X class itself

MRO:Method Resolution order : MRO Represents how properties(Attributes+Methods) are search in inheritance
 
* rule 1 - Python first search in child class and then goes to parent class , Priority is child class 
* rule 2 - MRO follow 'Depth first left to right Approach' 
* rule 3 - You can use mro() method for knowing mro of any class bject

```python
class a:
    pass

class b:
    pass

class c:
    pass

class x(a,b,c):
    pass

class y(b,c):
    pass

class p(x,y):
    pass

p.mro()
```

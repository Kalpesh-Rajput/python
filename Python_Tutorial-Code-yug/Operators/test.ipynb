```python
def fib(n):
    if n==1:
        return 0
    if n==2:
        return 1
    
    return (fib(n-2)+fib(n-1))


fib(10)
```

```python
def decor(func):
    def inner(*args):
        for num in args[1:]:
            if num == 0:
                return "cant divide"
        return func(*args)
    return inner


@decor
def div1(a,b):
    return a/b

@decor
def div2(a,b,c):
    return a/b/c





print(div1(10,0))
print(div1(0,10))
print(div1(20,15))
print(div2(10,0,20))
print(div2(0,10,20))
print(div2(15,10,20))
```
